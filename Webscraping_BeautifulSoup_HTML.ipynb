```python
# Download required modules
from bs4 import BeautifulSoup 
import requests 
import html5lib
import lxml
import pandas as pd
```

## Using BeautifulSoup object for scraping links and images

```python
# Store url
url_1 = "http://www.ibm.com"
```

```python
# Download webpage content in text format
data_1  = requests.get(url_1).text 
```

```python
# Create a BeautifulSoup object for the variable "data"
soup_1 = BeautifulSoup(data_1,"html.parser")
```

```python
# Use a for loop to scrape all links from the webpage (<a> tag)

for link in soup_1.find_all('a',href=True):

    print(link.get('href'))
```

```python
# Use a for loop to scrape all images from the webpage (<img> tag)
for link in soup_1.find_all('img'):
    print(link)
    print(link.get('src'))
```

## Using BeautifulSoup object for scraping an HTML table

```python
# Store url
url_2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
```

```python
# Download webpage content in text format
data_2  = requests.get(url_2).text
```

```python
# Create a BeautifulSoup object for the variable "data_2"
soup_2 = BeautifulSoup(data_2, "html.parser")
```

```python
# Find the HTML table on the webpage (<table> tag)
table = soup_2.find('table') 
```

```python
# Use a for loop to retrieve all rows and columns (<tr> tag for rows and <td> tag for columns) from the table
for row in table.find_all('tr'): 
    cols = row.find_all('td') 
    color_name = cols[2].string # re-name col #3
    color_code = cols[3].string # re-name col #4
    print("{}--->{}".format(color_name,color_code))
```

## Scraping HTML table into a DataFrame using read-html

```python
# Store url
url_3 = "https://en.wikipedia.org/wiki/World_population"
```

```python
dataframe_list = pd.read_html(url_3, flavor='bs4')
len(dataframe_list)
```

```python
# Select table # 5 
dataframe_list[5]
```

```python
# OR
pd.read_html(url_3, match="10 most densely populated countries", flavor='bs4')[0]
```
